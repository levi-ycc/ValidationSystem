```python
import fbm
import matplotlib.pyplot as plt
import numpy as np
```

# Fractional Brownian Motion

```python
f = fbm.FBM(n=1024, hurst=0.5, length=1, method='daviesharte').fbm()
```

```python
plt.plot(f)
```

```python
lowest = 0
highest = 0
u_l = ['None' for i in range(f.shape[0])]
window = 10
for n in range(0, f.shape[0]-window, window):
    for nn in range(window):
        if n+nn >= f.shape[0]:
            break
        if f[n+nn] > highest:
            u_l[n+nn] = 'upward'
        elif f[n+nn] < lowest:
            u_l[n+nn] = 'downward'
        else:
            u_l[n+nn] = 'remain'
    
    if n+window < f.shape[0]:
        lowest = np.min(f[n:n+window])
        highest = np.max(f[n:n+window])
```

```python
len(u_l)
```
