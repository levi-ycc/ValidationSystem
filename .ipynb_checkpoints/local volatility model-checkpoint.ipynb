```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def generate_lvm(n, periods, S0, r, q, sigma):
    series = np.zeros((n, periods+1))
    for i in range(n):
        w = [np.random.normal(0, 1/periods) for a in range(periods)]
        t=1/periods
        randw = [(r-q) * t + sigma*w[i] for i in range(len(w))]
        ds = S0*randw
        df = pd.DataFrame(ds, columns=['price'])
        df = pd.concat([pd.DataFrame([S0], columns=['price']), df], ignore_index=True)
        series[i] = df['price'].cumsum().values
    return series
```

```python
S0=1
```

```python
r=0.1
```

```python
q=0.1
```

```python
sigma=0.3
```

```python
series = generate_lvm(10,252, S0, r, q, sigma)
```

```python
plt.plot(series[3])
```
